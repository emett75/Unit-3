Before we can model, we will need to make some changes to the data set to allow to do so.

```python
#convert height to a number

def convert_Height(item):
    a = int(item.split('-')[0])
    b = int(item.split('-')[1])
    h = a * 30.48 + b * 2.54
    return h

# convert height with cm into an int

def remove_cm(item):
    h = ''.join(list(item)[:-2])
    return int(h)

#convert kg to lb

def convert_to_lb(item):
    h = "".join(list(item)[:-2])
    a = int(h)/0.45359237
    return a
```

```python
# create a new feature for height

height_cm=df['Height'].apply(lambda x: convert_Height(x) if '-' in x else remove_cm(x))
```

```python
df['height_cm'] = height_cm
```

```python
new_position.value_counts()
```

Now, we will model this data set to see if we can accurately predict a player's position using height and weight.
First, we will use the random forest model.

```python
x = df[['height_cm', 'Weight']]
y = df['Position']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.33, random_state = 42)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators = 20)
rfc.fit(x_train, y_train)
pre = rfc.predict(x_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(classification_report(pre, y_test))
print('\n')
print(confusion_matrix(pre, y_test))
```
